# Keras 기본 예제(선형회귀 분석)
### Y = AX + B

```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

### 데이터 만들기
### Y = 2X

```python
# 데이터 저장 배열
input_x = []
output_y = []

#그래프 그리기
for x in range(1,10):
    input_x.append([x])
    output_y.append([x*2 + random.random()])

input_x = np.array(input_x)
output_y = np.array(output_y)

print(input_x)
print(output_y)
```

```python
plt.figure(figsize=(12,8))
plt.plot(input_x, output_y)
plt.grid()
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

### 모델 그리기
### 1 Layer, 1 dense
### ![이미지](http://tykimos.github.io/warehouse/2017-8-12-Numerical_Prediction_Model_Recipe_1m.png)

```python
model = Sequential()
# 1 layer, 1 dense
model.add(Dense(1,input_dim=1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

### 학습하기

```python
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='loss') # 조기종료 콜백함수 정의
history = model.fit(input_x, output_y, epochs=500, verbose=2, callbacks=[early_stopping])
```

## 학습결과 출력

```python
def result(model, history):
    # summarize history for loss
    plt.figure(figsize=(12,8))
    plt.plot(history.history['loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    plt.show()

    plt.figure(figsize=(12,8))
    plt.plot(input_x, model.predict(input_x))
    plt.grid()
    plt.show()
```

```python
result(model, history)
```

### 모델 그리기 & 학습
### 2 layer, 64 dense -> 1
### ![이미지](http://tykimos.github.io/warehouse/2017-8-12-Numerical_Prediction_Model_Recipe_2m.png)

```python
model = Sequential()
# 2 layer, 64 dense -> 1
model.add(Dense(64,input_dim=1, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
early_stopping = EarlyStopping(monitor='loss') # 조기종료 콜백함수 정의
history = model.fit(input_x, output_y, epochs=500, verbose=2, callbacks=[early_stopping])
```

```python
result(model, history)
```

### 모델 그리기 & 학습
### 3 layer, 64 dense -> 64 dense -> 1
### ![이미지](http://tykimos.github.io/warehouse/2017-8-12-Numerical_Prediction_Model_Recipe_3m.png)

```python
model = Sequential()
# 3 layer, 64 dense -> 64 dense -> 1
model.add(Dense(64,input_dim=1, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
model.summary()

early_stopping = EarlyStopping(monitor='loss') # 조기종료 콜백함수 정의
history = model.fit(input_x, output_y, epochs=500, verbose=2, callbacks=[early_stopping])
```

```python
# 학습결과 출력
result(model, history)
```
